```python
error_attend = []
attend_files = os.listdir('./1820유관중')

attend_csvs = [f for f in attend_files if f.endswith('csv')]
recent_attend = pd.read_csv('./1820유관중\\{}'.format(attend_csvs[0]))
for i,c in enumerate(attend_csvs):
    if i == 0:
        continue
    try:
        df = pd.read_csv('./1820유관중\\{}'.format(c),error_bad_lines=False)     
    except:
        df = pd.read_csv('./1820유관중\\{}'.format(c),error_bad_lines=False, encoding = 'cp1252')
    recent_attend = pd.concat([recent_attend,df])
recent_attend.head()
```

```python
attend.dropna(inplace = True)
absence.dropna(inplace= True)
recent_attend.dropna(inplace = True)
```

```python
attend['Date'] = pd.to_datetime(attend.Date.apply(lambda x: x.split('/')[0]+'/'+x.split('/')[1]+'/20'+x.split('/')[2][-2:]))
recent_attend['Date'] = pd.to_datetime(recent_attend.Date.apply(lambda x: x.split('/')[0]+'/'+x.split('/')[1]+'/20'+x.split('/')[2][-2:]))
```

```python
def winpoint(string):
    if string == 'H':
        return 1
    elif string =='D':
        return 0.5
    elif string =='A':
        return 0
    else : return 0
```

```python
attend['winpoint'] = attend['FTR'].apply(winpoint)
absence['winpoint'] = absence['FTR'].apply(winpoint)
recent_attend['winpoint'] = recent_attend['FTR'].apply(winpoint)
```

```python
def get_mean_df(attend):
    whole_attend_mean = attend.groupby('Div').mean()
    whole_attend_mean['AFR'] = whole_attend_mean['AF']/ (whole_attend_mean['HF']+whole_attend_mean['AF'])
    whole_attend_mean['ACR'] = (whole_attend_mean['AY']+whole_attend_mean['AR'])/ (whole_attend_mean['HR']+whole_attend_mean['HY']+whole_attend_mean['AY']+whole_attend_mean['AR'])
    return whole_attend_mean

def get_mean_year(attend):
    attend['year']=pd.DatetimeIndex(attend['Date']).year
    whole_attend_mean = attend.groupby('year').mean()
    whole_attend_mean['AFR'] = whole_attend_mean['AF']/ (whole_attend_mean['HF']+whole_attend_mean['AF'])
    whole_attend_mean['ACR'] = (whole_attend_mean['AY']+whole_attend_mean['AR'])/ (whole_attend_mean['HR']+whole_attend_mean['HY']+whole_attend_mean['AY']+whole_attend_mean['AR'])
    return whole_attend_mean
```

```python
#winpoint = 홈팀 승률, AFR = 어웨이 파울 비율, #ACR = 어웨이 카드 비율
```

```python
#전체기간 유관중 평균
whole_attend_mean = get_mean_df(attend)
whole_attend_mean.to_excel('전체기간유관중평균.xlsx')
whole_attend_mean
```

```python
df = pd.concat([absence,recent_attend])

home_attend =recent_attend[['HomeTeam','FTHG','HS','FTR','HF','HY','HR','winpoint']].rename(columns = {'HomeTeam':'Team'})
away_attend = recent_attend[['AwayTeam','FTAG','AS','AST','AF','AY','AR','winpoint']].rename(columns = {'AwayTeam':'Team'})
home_absent = absence[['HomeTeam','FTHG','HS','FTR','HF','HY','HR','winpoint']].rename(columns = {'HomeTeam':'Team'})
away_absent = absence[['AwayTeam','FTAG','AS','AST','AF','AY','AR','winpoint']].rename(columns = {'AwayTeam':'Team'})
```

```python
def more_10_games(home_attend):
    temp = home_attend.groupby('Team').count()
    return home_attend[home_attend.Team.isin(temp[temp > 10].index)]
```

```python
away_attend= more_10_games(away_attend)
home_attend = more_10_games(home_attend)
home_absent= more_10_games(home_absent)
away_absent= more_10_games(away_absent)
```

```python
def reverse_winpoint(number):
    if number ==1:
        return 0
    elif number ==0:
        return 1
    elif number ==0.5:
        return 0.5
```

```python
away_attend['winpoint']=away_attend['winpoint'].apply(reverse_winpoint)
away_absent['winpoint']=away_absent['winpoint'].apply(reverse_winpoint)
```

```python
def get_mean(away_attend,home_attend):
    attend_team_sum = away_attend.groupby('Team').sum().merge(home_attend.groupby('Team').sum(),on='Team')
    attend_team_count = away_attend.groupby('Team').count().merge(home_attend.groupby('Team').count(),on='Team')
    attend_team_sum['winpoint'] = attend_team_sum['winpoint_x'] + attend_team_sum['winpoint_y']
    attend_team_count['winpoint'] = attend_team_count['winpoint_x'] + attend_team_count['winpoint_y']
    attend_team_mean = attend_team_sum/attend_team_count
    attend_team_mean.rename(columns={'winpoint_x':'원정승률','winpoint_y':'홈승률','winpoint':'총승률'},inplace = True)
    return attend_team_mean
```

```python
attend_team_mean = get_mean(away_attend,home_attend).dropna(axis=1)
absent_team_mean = get_mean(away_absent,home_absent).dropna(axis=1)
```

```python
attend_team_mean.to_excel('팀 유관중 평균.xlsx')
absent_team_mean.to_excel('팀 무관중 평균.xlsx')
```

```python
team_differ=absent_team_mean - attend_team_mean
team_differ_rate = (team_differ*100/attend_team_mean).dropna(axis=0)
team_differ_rate['총승률value차']=team_differ['총승률']
team_differ_rate['원정승률value차']=team_differ['원정승률']
team_differ_rate['홈승률value차']=team_differ['홈승률']
team_differ_rate.to_excel('팀 유무 차이.xlsx')
```
